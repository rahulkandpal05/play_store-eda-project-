## I passes a list ,in which only int and float datatypes are included

```python
num=[]
for i in df.columns:
    if df[i].dtype=='int32' or df[i].dtype=='float64':
        num.append(i)
```

## outliers

## now , i am passing a loop in which only numercial values included , in which i am creating  a Boxplot.

```python
for i in num:
    sns.boxplot(df[i])
    plt.show()
```

## I am creating a loop , in which i am calculating percentile(75,25),IQR,Lower_fence,Higher_fence. 
## And, i printed lower_fence  and higher_fence. 

```python
for i in num:
    q75,q25 = np.percentile(df[i] , [75,25])
    iqr=q75-q25
    lower_fence = q25-1.5*(iqr)
    higher_fence = q75+1.5*(iqr)
    print(i,'lower_limit :',lower_fence ,'upper_limit :',higher_fence)
```

## use capping method

```python
df['Rating']=np.where(df['Rating']>5.25,5.25,
               np.where(df['Rating']<3.25,3.25,
                       df['Rating'])      
                               )
df['Reviews']=np.where(df['Reviews']>73585.625,73585.625,
               np.where(df['Reviews']<-44111.375,-44111.375,
                       df['Reviews'])      
                               )
df['Size']=np.where(df['Size']>62490.7,62490.7,
               np.where(df['Size']<-37484.5,-37484.5,
                       df['Size'])      
                               )
df['Installs']=np.where(df['Installs']>2498500,2498500,
               np.where(df['Installs']<-1497500,-1497500,
                       df['Installs'])      
                               )
```

## now , i check the loop and i found that all the outliers are removed

```python
for i in num:
    sns.boxplot(df[i])
    plt.show()
```
